```python
import pandas as pd
data = pd.read_stata("C:/Users/oscar/OneDrive/Desktop/Dissertation/Stata File/panel with region as dummy with comparison.dta")   

#Load data to Stata
from pystata import stata
stata.pdataframe_to_data(data,True)
```

```python
#data description
#data summary
stata.run('''
describe
ssc install xtabond2
''',quietly=True)
```

```python
#declare individual identifier and time indentifier
stata.run('''
xtset id year
''',quietly=True)
#panel description of dataset
stata.run('''
xtdescribe
''',quietly=False)
```

```python
# panel summary statistics (within and between variation)
stata.run('''
xtsum id year rank gdp pollctl popu SO2 wage fdi north northeastern east center south southwestern northwestern
''',quietly=False)
```

```python
# within or FE and RE estimator with cluster-robust standard effors.
stata.run('''
xtreg fdi pollctl popu SO2 wage gdp rank, fe vce(cluster id)
xtreg fdi pollctl popu SO2 wage gdp rank, re vce(cluster id)
xtivreg fdi pollctl popu SO2 wage gdp (rank=north northeastern east center south southwestern), fe
xtivreg fdi pollctl popu SO2 wage rank (gdp=north northeastern east center south southwestern), re
''',quietly=False)
```

```python
# test re or fe
stata.run('''
xtreg lfdi lgovcrime lSO2 lpopu lwage rank north northeastern east center south southwestern, re cluster (id)
xtreg lfdi lgovcrime lSO2 lpopu lwage rank north northeastern east center south southwestern, fe cluster (id)
. qui xtreg lfdi lgovcrime lSO2 lpopu lwage rank north northeastern east center south southwestern, re
. estimates store re
. qui xtreg lfdi lgovcrime lSO2 lpopu lwage rank north northeastern east center south southwestern, fe
. estimates store fe
''',quietly=False)
```

#### CONCLUSION: FE is better

### test of time effect

```python
# wald test
stata.run('''
xtreg lfdi lgovcrime lSO2, fe
// test year2 = year3 = year4 = year5 = year6 = year7= year8 = year9 = year10//////////////////////////year2 not found
''',quietly=False)
```

```python
# LM test
stata.run('''
xtreg lfdi lgovcrime lSO2 , fe
est store fe
xtreg lfdi lgovcrime lSO2 year*, fe
est store fe_dum
lrtest fe fe_dum
''', quietly=False)
#year dummy is effective
```

```python
# comparison between fe_rebust fe_id fe_region
stata.run('''
 xtreg lfdi lgovcrime lSO2 lpollctl lwage lpopu rank, fe robust
 est store fe_rb
 xtreg lfdi lgovcrime lSO2 lpollctl lwage lpopu rank, fe cluster(id)
 est store fe_cid
 xtreg lfdi lgovcrime lSO2 lpollctl lwage lpopu rank, fe cluster(id)
 est store fe_cregion
 local models "fe_rb fe_cid fe_cregion"
 estimate table `models'
 ''',quietly=False)
```

### IV selection

```python
# region as dummy
stata.run('''
 xtivreg lfdi lgovcrime (lpollctl=east center south southwestern) lSO2 lwage lpopu rank, fe
 est store fe_ivregion
 xtreg lfdi lgovcrime lpollctl lSO2 lwage lpopu rank, fe
 est store fe
''',quietly=False)
```

```python
# rank as dummy
stata.run('''
xtivreg lfdi (lgovcrime=rank)  lpollctl lSO2 lwage lpopu, fe
est store fe_01
xtreg lfdi lgovcrime rank lpollctl lSO2 lwage lpopu, fe
est store fe_ivrank
local models "fe_01 fe_ivrank"
est table `models'
''',quietly=False)

# Reject null hypothesis
# Linear is better
```
